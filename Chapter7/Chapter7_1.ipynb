```python
#07-02

import torch
import torch.nn as nn
```

```python
# 배치 크기 x 채널 x 높이 x 너비
inputs = torch.Tensor(1, 1, 28, 28)
```

```python
#1채널 -> 32채널, 커널사이즈 3
conv1 = nn.Conv2d(1, 32, 3, padding=1)
print(conv1)
```

```python
#32채널 -> 64채널, 커널사이즈 3, 패딩 1
conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
print(conv2)
```

```python
pool = nn.MaxPool2d(2)
print(pool)
```

```python
out = conv1(inputs)
print(out.shape)
```

```python
out = pool(out)
print(out.shape)
```

```python
out = conv2(out)
print(out.shape)
```

```python
out = pool(out)
print(out.shape)
```

```python
#첫번째 차원인 배치 차원은 그대로 두고 나머지는 펼친다
out = out.view(out.size(0),-1)
print(out.shape)
```

```python
fc = nn.Linear(3160, 10)
out = fc(out)
print(out.shape)
```
